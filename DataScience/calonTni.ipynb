```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('calon_tni.csv')
tertinggi = np.max(data['tinggi'])
terendah = np.min(data['tinggi'])
medianTinggi = np.median(data['tinggi'])
meanTinggi = np.mean(data['tinggi'])
```

```python
print("=== Data Calon TNI ===" )
print(data)
```

```python
print("\n=== Statistik Dasar Tinggi Badan ===") 
print(data["tinggi"].describe())
print("\nRata-rata tinggi:", meanTinggi)
print("Tertinggi:", tertinggi)
print("Terendah:", terendah)
print("Median:", medianTinggi)
```

```python
listTertinggi = data[data['tinggi'] > meanTinggi]

print("=== Calon di atas rata-rata tinggi ===")
print(listTertinggi)
```

```python
tertinggiSorted = data.sort_values(by="tinggi", ascending=False)

print("\n=== Urutan Dari tertinggi hingga terendah ===")
print(tertinggiSorted)
```

```python
listKategori = {
    "nama": [],
    "alamat": [],
    "tinggi": [],
    "kategori": []
}
for index, row in data.iterrows():
    if row["tinggi"] >= 180:
        listKategori["nama"].append(row["nama"])
        listKategori["alamat"].append(row["alamat"])
        listKategori["tinggi"].append(row["tinggi"])
        listKategori["kategori"].append("Tinggi")
    elif row["tinggi"] >= 170:
        listKategori["nama"].append(row["nama"])
        listKategori["alamat"].append(row["alamat"])
        listKategori["tinggi"].append(row["tinggi"])
        listKategori["kategori"].append("Sedang")
    else:
        listKategori["nama"].append(row["nama"])
        listKategori["alamat"].append(row["alamat"])
        listKategori["tinggi"].append(row["tinggi"])
        listKategori["kategori"].append("Pendek")

print("\n=== Kategori Tinggi Badan ===")
print(pd.DataFrame(listKategori))
```

```python
kategoriCounts = pd.Series(listKategori["kategori"]).value_counts()

print("=== Jumlah calon per kategori ===")
print(kategoriCounts)
```

```python
print("=== Statistika lanjutan===")
print("Standar Deviasi:", np.std(data["tinggi"]).round(2))
print("Variansi:", np.var(data["tinggi"]).round(2))
```

```python


print("=== Calon dengan tinggi badan tertinggi ===")
calonTertinggi = data[data["tinggi"] == tertinggi].values.tolist()
print("Nama:", calonTertinggi[0][0])
print("Alamat:", calonTertinggi[0][1])
print("Tinggi:", calonTertinggi[0][2])
```
