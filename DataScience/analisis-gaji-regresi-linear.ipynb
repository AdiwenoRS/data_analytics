```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data
df = pd.read_csv('Salary_Data.csv')
X = df[['YearsExperience']]  # DataFrame dengan kolom 'YearsExperience'
y = df['Salary']

# Latih model
model = LinearRegression()
model.fit(X, y)

# Visualisasi
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title('Gaji vs Pengalaman Kerja')
plt.xlabel('Pengalaman (tahun)')
plt.ylabel('Gaji ($)')
plt.show()

# Prediksi dengan konsistensi nama kolom
new_data = pd.DataFrame({'YearsExperience': [5]})
prediksi = model.predict(new_data)[0]
print("Prediksi gaji untuk 5 tahun pengalaman:", prediksi)
```
