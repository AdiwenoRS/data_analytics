```python
import pandas as pd
import numpy as np  
import scipy as sp
```

```python
a = [10, 3, 5, 12, 50]
b = [5, 4, 3, 2, 1]
array_a = np.array(a)
array_b = np.array(b)
```

```python
print(a)
print(b)
print(array_a)
print(array_b)

print("multiplication:", array_a * array_b)
print("addition:", array_a + array_b)
print("subtraction:", array_a - array_b)
print("division:", array_a / array_b)
```
