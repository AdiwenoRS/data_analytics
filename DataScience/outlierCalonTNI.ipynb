```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("calon_tni2.csv")
print(df)
```

```python
Q1 = df['gaji'].quantile(0.25)
Q3 = df['gaji'].quantile(0.75)
IQR = Q3 - Q1

outlier = (Q3 + 1.5 * IQR)
gajiOutlier = df[
    (df['gaji'] > outlier)
]
```

```python
print(outlier)
print(gajiOutlier)
print(f"Jumlah outlier: {gajiOutlier.shape[0]}")
print(gajiOutlier[['nama', 'gaji']])
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(df['gaji'] / 1000000)
plt.ylabel('gaji (dalam juta rupiah)')
plt.grid(True, alpha=0.3)
plt.show()
```
