```python
class Stack:
      def __init__ (self):
            self.items = [] # Variable of list
            
      # Menambah elemen ke atas tumpukan
      def push(self, item):
            self.items.append(item)
            
      # Menghapus dan mengembalikan elemen paling atas
      def pop(self):
            if not self.is_empty():
                  return self.items.pop() # Metode pop() tanpa argumen akan menghapus data paling atas
            else:
                  return "Stack Kosong"
      
      # Melihat elemen paling atas tanpa menghapus
      def peek(self):
            if not self.is_empty():
                  return self.items[-1]
            else:
                  return "Stack Kosong"
      
      # Mengecek apakah tumpukan kosong
      def is_empty(self):
            return len(self.items) == 0
      
      # Menampilkan isi tumpukan
      def display (self):
            print("Isi Stack (Top -> Bottom): ", list(reversed(self.items)))
            
      # Menghitung jumlah elemen dalam tumpukan
      def size(self):
            return len(self.items)
            
```

```python
# Contoh penggunaan

stack = Stack() # define the class

for i in range(10):
      stack.push(i)

stack.display()  # Output: Isi Stack (Top -> Bottom): [Values]

print(stack.pop())  # Output

print(f"Jumlah elemen dalam stack: {stack.size()}") 

stack.display()  # Output: Isi Stack (Top -> Bottom): [Values]

print(f"Jumlah elemen dalam stack: {stack.size()}") 

print(stack.peek())  # Output
```
